```python
# Initializing imports.
import pandas as pd 
from pathlib import Path
import datetime as dt
```

```python
# Read in the data that was copied to our clipboard and converting it to a dataframe.
Cubs_Games_2019 = pd.read_clipboard()
Cubs_Games_2019.head()
```

```python
# Drop the columns that are not needed.
Cubs_Games_2019.drop(columns = ['Game','Score', 'Decision', 'Record'], inplace = True)
```

```python
Cubs_Games_2019.head()
```

```python
# Find Home Games only and excluding any away games.
Cubs_Home_Games_2019 = Cubs_Games_2019[Cubs_Games_2019.Opponent.str.contains('vs', regex=False)]
```

```python
Cubs_Home_Games_2019
```

```python
# Convert the dates into the datetime format.
Cubs_Home_Games_2019['Date / Box Score'] = pd.to_datetime(Cubs_Home_Games_2019['Date / Box Score'])
```

```python
Cubs_Home_Games_2019.head()
```

```python
# Get the number of the week for each date on daily dataframe.
Week_of_Cubs_Game = Cubs_Home_Games_2019['Date / Box Score'].dt.week.to_frame()
Week_of_Cubs_Game
```

```python
# Merge the week game data with the rest of the dataframe. Each date now has its respective week number.
Cubs_Weekly_Home_Games = pd.concat([Week_of_Cubs_Game, Cubs_Home_Games_2019], axis = 'columns', join = 'inner')
Cubs_Weekly_Home_Games
```

```python
# Rename column names.
columns = ['Week', 'Date / Box Score', 'Opponent']
Cubs_Weekly_Home_Games.columns = columns
Cubs_Weekly_Home_Games.head()
```

```python
# Group dataframe by 'Week' number and count the number of games each week, then sort them by ascending date. 
Cubs_Weekly_Home_Games.groupby('Week').count()
```

```python
# Rename the games column and dropp the 'Opponent' column.
Cubs_2019_Home_Games = Cubs_Weekly_Home_Games.rename(columns={'Date / Box Score':'No_Cubs_Games'}).drop(columns=['Opponent'])
```
